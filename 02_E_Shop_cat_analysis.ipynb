```python
#𝗣𝗿𝗼𝗷𝗲𝗰𝘁 𝗤𝘂𝗲𝘀𝘁𝗶𝗼𝗻 # 𝟮 – 𝗘-𝗦𝗵𝗼𝗽 𝗖𝗮𝗿𝘁 𝗔𝗻𝗮𝗹𝘆𝘀𝗶𝘀

#🧠 Develop a program that:
#Takes input for 5 items: name, price, quantity (as strings)
#Convert all values using casting to int and float
#Store data in lists and dictionaries
#Add 10% tax, calculate subtotal, total, and average
#Identify high-value items (price > 500) and store in a set, convert to frozenset
#Use functions: calculate_bill(), apply_discount(), print_cart_summary()
#Use loops to display item info with type() checks
#Use random.choice() for a giveaway item
#Use keyword module to verify if item name is a Python keyword
#Use datetime to log purchase time and calendar to show current month


import random
import keyword
import datetime
import calendar
store_dict = {}
sub_total = 0
item_quantity = 0
high_value_items = set()
def calculate_bill(price, quantity):
    return price * quantity
def apply_discount(total):
    return total + total / 10
def print_cart_summary(store, total_units, total_price):
    for item, info in store.items():
        print("\n ------------ \n")
        print(f"Item: {item}", type(item))
        print(f"Is name of item keyword? {keyword.iskeyword(item)}")
        print(f"Price: {info['price'][0]}", type(info["price"][0]))
        print(f"Quantity: {info['quantity']}", type(info["quantity"]))
        print(f"Item total: {info['item_total'][0]}", type(info["item_total"][0]))
    print("\n ------------ \n")
    print(f"Total units: {item_quantity}", type(item_quantity))
    print(f"Subtotal: {sub_total}", type(sub_total))
    grand_total = apply_discount(sub_total)
    print(f"Grand total: {grand_total}", type(grand_total))
    average = grand_total / item_quantity
    print(f"Average per unit item: {average}", type(average))
    giveaway_item = random.choice(list(store.keys()))
    print(f"Random giveaway item: {giveaway_item.upper()}", type(giveaway_item))
    print(f"Current date and time: {datetime.datetime.now()}")
    current_month_number = datetime.datetime.now().month
    current_month_name = calendar.month_name[current_month_number]
    print(f"Current month: {current_month_name}")


for i in range(5):
    item = input(f"Enter the name of item {i + 1}: ")
    str_price = input("Enter its price: ")
    str_quantity = input("Enter its quantity: ")
    float_price = float(str_price)
    int_quantity = int(str_quantity)
    item_total = calculate_bill(float_price, int_quantity)
    prices= []
    prices.append(float_price)
    item_total_list = []
    item_total_list.append(item_total)
    store_dict[item] = {"price": prices, "quantity": int_quantity, "item_total": item_total_list}
    sub_total += item_total
    item_quantity += int_quantity
    if store_dict[item]["price"][0] > 500:
        high_value_items.add(item)
grand_total = apply_discount(sub_total)
average = grand_total / item_quantity
fix_high_value_items = frozenset(high_value_items)
print("\n")
print(f"High value items: {fix_high_value_items}", type(fix_high_value_items))
print("\n")
print_cart_summary(store_dict, item_quantity, sub_total)
```
