```python
#𝗣𝗿𝗼𝗷𝗲𝗰𝘁 𝗤𝘂𝗲𝘀𝘁𝗶𝗼𝗻 # 𝟯 – 𝗦𝗺𝗮𝗿𝘁 𝗔𝗱𝗱𝗿𝗲𝘀𝘀 𝗕𝗼𝗼𝗸

#🧠 Design a system that:
#Takes input for contacts: name, phone, age, city
#Store in a list of dictionaries
#Use type casting for age and phone
#Create functions: add_contact(), search_contact(), display_all()
#Use conditions to find contacts from specific cities or age > 30
#Check name validity using keyword module
#Use os.getcwd() to show file location
#Convert data to a tuple, then frozenset
#Display each contact's data type
#Log time of contact addition using datetime


import os
import datetime
import keyword
over_30_contacts = []
gujranwala_contacts = []
contacts = []
def add_contact():
    contact_dict = {}
    name = input("Enter name here: ").lower()
    str_phone = input("Enter phone number here: ")
    str_age = input("Enter age here: ")
    city = input("Enter city here: ").lower()
    phone = int(str_phone)
    age = int(str_age)
    contact_dict = {"name": name, "age": age, "phone": phone, "city": city}
    contacts.append(contact_dict)
    print("Contact added at ", datetime.datetime.now())
def search_contact():
    search_key = input("Will you search by name or city? ").lower()
    search_value = input("Enter the name or city to search: ").lower()
    for contact in contacts:
        if contact[search_key].lower() == search_value:
            print("Contact found!")
            print(contact)
            break
        else:
            print("Contact not found")
all_contact_tuple = []
all_contact_frozenset = []
def display_all():
    for contact in contacts:
        print("\n ----------------\n")
        print(contact)
        print("Is name of contact keyword? ", keyword.iskeyword(contact["name"]))
add_contact()
add_contact()
search_contact()
display_all()
for contact in contacts:
    if contact["city"].lower() == "gujranwala" and contact["age"] > 30:
        gujranwala_contacts.append(contact)
        over_30_contacts.append(contact)
    elif contact["city"].lower() == "gujranwala":
        gujranwala_contacts.append(contact)
    elif contact["age"] > 30:
        over_30_contacts.append(contact)
    contact_tuple = tuple(contact.values())
    contact_frozenset = frozenset(contact_tuple)
    all_contact_tuple.append(contact_tuple)
    all_contact_frozenset.append(contact_frozenset)

for x in all_contact_tuple:
    for y in x:
        print(y, type(y))

print("Gujranwala contacts: ", gujranwala_contacts)
print("Contacts with age over 30 years: ", over_30_contacts)
print("Current file: ", os.getcwd())
```
