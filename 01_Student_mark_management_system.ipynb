```python

#ð—£ð—¿ð—¼ð—·ð—²ð—°ð˜ ð—¤ð˜‚ð—²ð˜€ð˜ð—¶ð—¼ð—» # ðŸ­ â€“ ð—¦ð˜ð˜‚ð—±ð—²ð—»ð˜ ð— ð—®ð—¿ð—¸ ð— ð—®ð—»ð—®ð—´ð—²ð—ºð—²ð—»ð˜ ð—¦ð˜†ð˜€ð˜ð—²ð—º
#ðŸ§  Create a Python program where you:
#Take input for 5 students (name, age, city, and 3 subject scores as strings)
#Store data using lists, tuples, and dictionaries
#Use type casting to convert scores to integers
#Calculate total, average (using statistics.mean()), and grade using if-elif-else
#Define functions: calculate_grade(), is_passed(), display_student_info()
#Store results in a set, remove duplicates, and convert to frozenset
#Use datetime to log current report time
#Use math to find square root of total marks
#Use keyword to check if names are Python keywords
#Display data types using type()
#Use os to print current working directory
#Print final results using loops and conditions
result_set = set()
from statistics import mean
import datetime
import math
import keyword
import os
data_dict = {}
for i in range(5):
    names = input(f"Enter the name of student {i + 1}: ")
    age = int(input("Enter his age: "))
    city = input("Enter his city: ")
    scores_list_str = []
    scores_list_int = []
    for j in range(3):
        str_scores = input(f"Enter marks of subject {j + 1}: ")
        scores_list_str.append(str_scores)
        scores_list_int.append(int(str_scores))
    info_tuple = (names, age, city)
    total = sum(scores_list_int)
    average = mean(scores_list_int)
    data_dict[names] = {"info": info_tuple, "scores": scores_list_int, "total": total, "average": average}
print("Report time:")
print(datetime.datetime.now())
def calculate_grade(average):
    if average >= 90:
        return "Grade: A+"
    elif average >= 80:
        return "Grade: A"
    elif average>= 70:
        return "Grade: C"
    elif average >= 60:
        return "Grade: D"
    else:
        return "Fail"
def is_passed(average):
    if average >= 60:
        return "Yes, Passed"
    else:
        return "No, Failed"
def display_student_info(student):
    print("\n ----------- \n")
    name, age, city = student["info"]
    scores = student["scores"]
    total = student["total"]
    average = student["average"]
    grade = calculate_grade(student["average"])
    passed = is_passed(student["average"])
    print(f"Name: {name}", type(name))
    print(f"Age: {age}", type(age))
    print(f"City: {city}", type(city))
    print(f"Scores: {scores}", type(scores))
    print(f"Total: {total}", type(total))
    print(f"Average: {average}", type(average))
    print(f"{grade}", type(grade))
    print(f"Is passed: {passed}", type(passed))
for student in data_dict.values():
    display_student_info(student)
    grade = calculate_grade(student["average"])
    passed = is_passed(student["average"])
    total_sqrt = math.sqrt(student["total"])
    print(f"square root of total marks: {total_sqrt}", type(total_sqrt))
    is_name_keyword = keyword.iskeyword(name)
    print(f"Is name keyword? {is_name_keyword}", type(is_name_keyword))
    result_set.add((grade, passed))
unique_set = frozenset(result_set)
print(unique_set)
print(os.getcwd())
```
