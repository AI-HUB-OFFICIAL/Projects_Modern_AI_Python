```python
# 𝗣𝗿𝗼𝗷𝗲𝗰𝘁 𝗤𝘂𝗲𝘀𝘁𝗶𝗼𝗻 # 𝟱 – 𝗙𝘂𝗹𝗹 𝗙𝗲𝗮𝘁𝘂𝗿𝗲𝗱 𝗣𝗲𝗿𝘀𝗼𝗻𝗮𝗹 𝗣𝗿𝗼𝗳𝗶𝗹𝗲 𝗕𝘂𝗶𝗹𝗱𝗲𝗿
#🧠 Write a program that:
#Takes input: name, age, height, skills (list), hobbies (set), and profile summary (multi-line string)
#Convert data using casting where needed
#Use math to calculate BMI, datetime for birth year, os for current directory
#Store all info in a dictionary and display with a loop
#Check if any skills are Python keywords
#Create functions: calculate_bmi(), profile_summary(), display_info()
#Convert skills to frozenset and hobbies to list
#Use conditions to classify user as Beginner, Intermediate, or Expert
#Use calendar.month() to show joining month



import math
import datetime
import os
import keyword
import calendar
store_dict = {}
def calculate_bmi(weight, height):
    bmi = weight / math.pow(height, 2)
    return bmi
def profile_summary():
    print("Enter your profile summary in multi lines. Write 'END' when you're done.")
    lines = []
    while True:
        line = input()
        if line.strip().upper() == 'END':
            break
        lines.append(line)
    summary = "\n".join(lines)
    return summary
def display_info(store):
    name = store["name"]
    age = store["age"]
    height = store["height"]
    weight = store["weight"]
    bmi = store["bmi"]
    birthday_year = store["birthday_year"]
    joining_month = store["joining_month"]
    skills = store["skills"]
    classification = store["classification"]
    hobbies = store["hobbies"]
    profile_summary = store["profile_summary"]
    print("Name: ", name)
    print("Age: ", age)
    print("Height: ", height)
    print("Weight: ", weight)
    print("BMI: ", bmi)
    print("Birthday year: ", birthday_year)
    print("Joining month: ", joining_month)
    print("Skills: ", skills)
    print("Classification: ", classification)
    print("Hobbies: ", hobbies)
    print("Profile summary: ", profile_summary)
name = input("Enter your name: ")
str_age = input("Enter your age: ")
str_height = input("Enter your height(in meters): ")
str_weight = input("Enter your weight(in kilograms): ")
raw_skills = input("Enter your skills(comma seperated): ")
skills_list = [skill.strip() for skill in raw_skills.split(',')]
raw_hobbies = input("Enter your hobbies(comma seperated): ")
hobbies_set = set(hobby.strip() for hobby in raw_hobbies.split(','))
age = int(str_age)
height = float(str_height)
weight = float(str_weight)
bmi = calculate_bmi(weight, height)
current_year = datetime.datetime.now().year
birthday_year = current_year - age
joining_month_number = datetime.datetime.now().month
joining_month = calendar.month_name[joining_month_number]
if len(skills_list) <= 2:
    classification = "Beginner"
    store_dict["classification"] = classification
elif 3 <= len(skills_list) <= 4:
    classification = "Intermediate"
    store_dict["classification"] = classification
else:
    classification = "Expert"
    store_dict["classification"] = classification
skills_set = set(skills_list)
skills = frozenset(skills_set)
hobbies = list(hobbies_set)
profile_summary = profile_summary()
store_dict = {"name": name, "age": age, "height": height, "weight": weight, "bmi": bmi, "birthday_year": birthday_year, "joining_month": joining_month, "skills": skills, "classification": classification, "hobbies": hobbies, "profile_summary": profile_summary}
print("\n Profile info: ")
display_info(store_dict)
for skill in skills:
    print("Are skills keywords: ", keyword.iskeyword(skill))
print("Current working directory", os.getcwd())
```
